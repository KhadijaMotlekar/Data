# Salaries of different degrees

```python
import matplotlib.pyplot as plt
#importing the library matplotlib.pyplot as plt
import pandas as pd
#importing pandas library as pd
import numpy as np
#importing the library numpy as np
import builtins
dir(builtins)

deg = pd.read_csv('degrees-that-pay-back.csv')
#reading and saving a csv file for employment in this kernel
deg
#displaying the employment csv file
```

```python
deg.columns = ['degrees', 'startSalary', 'midSalary', 'change', 'mid10', 'mid20', 'mid75', 'mid90']
deg.head()
```

```python
dollar = ['startSalary', 'midSalary', 'mid10', 'mid20', 'mid75', 'mid90']

for x in dollar:
    deg[x] = deg[x].astype(str).str.replace("$", "")
    deg[x] = deg[x].str.replace(",", "")
    deg[x] = pd.to_numeric(deg[x])

deg.head()
```

```python
import seaborn as sns

sns.barplot(x='degrees', y='startSalary', data = deg)
plt.figure(figsize=(9,6))
plt.show()
```

```python
plt.figure(figsize=(8,12))
deg.plot(kind='bar', x='degrees', y='change')
plt.xticks(x='degrees',rotation='vertical')
plt.title('Change in start salary and mid salary (%)')
plt.ylabel('Change in %')
plt.legend()
plt.show()
```
